```stata
use data/scfstudentdebt2019, clear
```

```stata
xtile incomedecile=income [pw=wgt], n(10)
label var incomedecile "Catherine/Yannelis income deciles"
label define incomedecile 1 "0-10%" 2 "10-20%" 3 "20-30%" 4 "30-40%" 5 "40-50%" ///
6 "50-60%" 7 "60-70%" 8 "70-80%" 9 "80-90%" 10 "90-100%"
label values incomedecile incomedecile
```

```stata
qui {

gen selfparent=self_fedloans + parentalfedloans
    
foreach amt in 10000 50000 75000 {
gen selfcancel`amt'=`amt'
    replace selfcancel`amt'=selfparent if selfparent<`amt'

gen spousecancel`amt'=`amt'
replace spousecancel`amt'=spouse_fedloans if spouse_fedloans<`amt'
  
egen householdcancel`amt'=rowtotal(selfcancel`amt' spousecancel`amt')

gen percapcancel`amt'=householdcancel`amt'
    replace percapcancel`amt'=householdcancel`amt' / 2  if married==1 & householdcancel`amt'!=0
    }
} 
```

```stata
*** Percentage Decline in Mean Debt-to-Income Ratio, by Inc_cat
*create variable for all debt
qui gen alldebt=-1*(networth-asset)
qui gen debtinc_ratio= alldebt/income
qui gen debtinc_ratioafter= (alldebt-householdcancel50000)/income
qui gen debtinc_ratio_change = (debtinc_ratio - debtinc_ratioafter) / debtinc_ratio * 100
```

```stata
qui regress debtinc_ratio_change i.incomedecile##i.race if selfstudent!=1 & spousestudent!=1 [pw=wgt]

qui margins i.incomedecile#i.race if race!=5

marginsplot, ytitle(% REDUCTION IN DEBT-TO-INCOME RATIO) ///
scheme(dubois) ylabel(0 "0%" 10 "10%" 20 "20%" 30 "30%") ///
title(FIGURE 4: DEBT-TO-INCOME RATIO REDUCTION, size(medsmall)) legend(pos(12) ) xlabel(,alternate) ///
xscale(titlegap(large)) xtitle(INCOME DECILES) graphregion(margin(large)) ///
note(" " "Note: Data from Survey of Consumer Finances 2019. Estimates are based" ///
        "on $50,000 maximum for debt cancellation. Excludes debt of adult children" ///
        "living with parents. Excludes debt of households where respondent or" ///
        "partner is currently a student.")

graph export figures/f4_debttoincomereduction.pdf, replace
```
