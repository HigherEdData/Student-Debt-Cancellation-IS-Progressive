## Figure 2: All Households vs Only Borrowers

from "Student Debt Cancellation *Is* Progressive", a Roosevelt Institute research paper 

by Charlie Eaton, Adam Goldstein, Laura Hamilton, and Frederick Wherry

```stata
ssc install dubois
```

```stata
graph set svg fontface Tahoma
graph set eps fontface Tahoma
graph set ps fontface Tahoma
set scheme dubois
```

```stata
use data/scfstudentdebt2019, clear
```

```stata
qui xtile incomedecile=income [pw=wgt], n(10)
label var incomedecile "Catherine/Yannelis income deciles"
label define incomedecile 1 "0-10%" 2 "10-20%" 3 "20-30%" 4 "30-40%" 5 "40-50%" ///
6 "50-60%" 7 "60-70%" 8 "70-80%" 9 "80-90%" 10 "90-100%"
label values incomedecile incomedecile
qui gen borroweronly=percapfedloan if percapfedloan!=0
```

```stata
graph bar (mean) totfedloan_debt borroweronly [pw=wgt] ///
if selfstudent!=1 & spousestudent!=1, over(incomedecile, label(alternate)) ///
scheme(dubois) bar(1, lcolor("101 67 33") fcolor("101 67 33")) bar(2, lcolor("255 192 203") ///
fcolor("255 192 203")) b1title(" " "INCOME DECILES" " ", size(3)) ytitle(MEAN DEBT PER CAPITA) ///
ylabel(0 "$0" 10000 "$10K" 20000 "$20K" 30000 "$30K" 40000 "$40K") ///
title("FIGURE 2: MISLEADING ESTIMATES, CASE B", size(3.5) ///
 span) legend(pos(12) order(2 1) label(2 "borrowers only") label(1 "all households")) ///
note("Note: Data from Survey of Consumer Finances 2019. Excludes debts for" ///
    "households where respondent or partner is currently a student.", size(small) xoffset(0)) graphregion(margin(large))

qui graph export figures/f2_borrowervshousehould.pdf, replace
```
